### Part B- Normalize the data

```python
# import necessary libraries
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#load the data
Data = pd.read_csv('concrete_data.csv')
Data.head()
```

```python
Data.shape
```

```python
Data.isnull().sum()
```

```python
Data.describe()
```

```python
# split the data as input and target,Create input by dropping the target variable Strength.
Input = Data.drop('Strength',axis=1)
Input.shape
```

```python
# split the data as input and target,Create the target data which is Strength here.
Target = Data['Strength']
```

```python
#Normalize the input data
Input_norm = (Input-Input.mean())/Input.std()
Input_norm.head()
```

```python
# mention number of input variable
n_cols=Input_norm.shape[1]
n_cols
```

```python
i=1
list_MSE=[]  # For getting mean_squared_error values list for 50 random executions use a while loop
while i<=50:
    
    #split the data for training and testing
    x_train,x_test,y_train,y_test = train_test_split(Input_norm,Target,test_size=0.3,random_state=1)
    
    # Define the model
    def regression_model():
        model = Sequential()
        model.add(Dense(10,activation='relu',input_shape=(n_cols,))) # adding one hidden layer
        model.add(Dense(1)) # adding output layer
    
        model.compile(optimizer='adam',loss='mean_squared_error')
        return model
    
    model = regression_model() #use the model we created
    model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=50,verbose=2) #fitting the data into the model
    prediction = model.predict(x_test)
    
      # Find Mean Squared Error
    MSE = mean_squared_error(y_test,prediction)
    list_MSE.append(MSE)
    i=i+1
    
list_MSE
```

```python
NP_list_MSE = np.array(list_MSE) #convert the list into a numpy array
```

```python
NP_list_MSE.shape
```

```python
mean_MSE_PartB = NP_list_MSE.mean() #find mean of the mean squared error
print('mean of mean squared error %.2f'% mean_MSE_PartB)
```

```python
std_MSE = NP_list_MSE.std() #find mean of the standard deviation
print('standard deviation of mean squared error %.2f'% std_MSE)
```

```python
#Find the difference of Part B while comparing to Part A
mean_MSE_PartA = 293.29
mean_MSE_PartB = 376.38

Difference = mean_MSE_PartB-mean_MSE_PartA 
print('mean of Mean Squared Error of Part B is {} more than Part A'.format('%.2f'% Difference))
```
